2. Create a dictionary to store some information about yourself.  For example you might have key value pairs like:
- favorite ice cream & chocolate
- hair color & blond

````python
about_me = {
    "favorite_ice_cream": "chocolate",
    "hair_color": "black",
    "favorite_hobby": "AI Automation",
    "has_pets": True
}

about_me["favorite_programming_language"] = "Python"

for key, value in about_me.items(): # The for loop prints out every label and answer
    clean_label = key.replace("_", " ").title()
    print(f"{clean_label}: {value}")
````

6. Create two sets and find the intersection between them.

````python
set_a = {"apple", "banana", "cherry", "kiwi"}
set_b = {"cherry", "kiwi", "mango", "watermelon"}
set_a, set_b
````

````python
intersection_set = set_a.intersection(set_b)
intersection_set
````

7. Find another method to run on one or both of your sets.

````python
set_a = {"apple", "banana", "cherry", "kiwi", "mandarin"}
set_b = {"cherry", "kiwi", "mango", "watermelon", "guava"}
set_a, set_b
````

````python
union_set = set_a.union(set_b) # Find items that are in set_a AND in set_b
union_set
````

````python
difference_set = set_a.difference(set_b) # Find items that are in set_a BUT NOT in set_b
difference_set
````
