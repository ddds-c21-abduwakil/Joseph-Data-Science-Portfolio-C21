# Data Science Problem Definition


For each item develop a well-worded data science problem statement.  Your answer should include what type of data science problem this is ( supervised vs unsupervised ) and then whether it is a classification, regression, or clustering problem.  If it is a supervised problem, you should state what the target of the prediction is.

1. Zillow wants you to predict real estate sales prices. They have a large
database of 12 years of home features accompanied by the corresponding
sales prices.


**Statement:**

Utilize 12 years of historical home features and corresponding transaction data to build a predictive model that estimates the continuous sales price of a residential property.

**Type of Data Science Problem:**

Supervised Learning (dataset includes both input features and historical labeled outputs/sales prices).

**Target:**

The sale price of the home (a continuous numerical variable).


---

2. Amazon wants you to develop a code that upsells (add additional items to the
customer’s order) based on what items have been viewed and bought by that
customer over the past 2 years.



**Statement:**

Analyze 2 years of customer viewing and purchase history to identify patterns and relationships among products, grouping similar items or customer behaviors together to generate relevant product recommendations for upselling.

**Type of Data Science Problem:**

Unsupervised Learning (the system is discovering inherent patterns, associations, or groupings in historical user behavior without relying on a predefined, labeled correct/incorrect target variable).


---

3. MGM wants to know what features of a movie seem to increase the movie’s
box office income. They have a large database of previously released movies
accompanied by the resulting box office revenue.



**Statement:**

Data Science Problem Statement: Analyze historical movie data to identify and quantify the relationships between specific movie features (such as budget, genre, cast, and runtime) and the resulting box office revenue.

**Type of Data Science Problem:**

Supervised Learning (since the dataset contains known historical features alongside their corresponding labeled outcome/revenue data).

**Target:**

Box office revenue (a continuous numerical variable).



---

4. Spotify wants you to develop a method of predicting whether a customer is
about to terminate their service BEFORE they terminate so that they can offer
enticements to the customer to keep the service and not terminate (called
‘churn’)



**Statement:**

Data Science Problem Statement: Build a predictive model using customer engagement metrics, usage patterns, and billing history to determine whether a customer is likely to terminate their subscription service before the cancellation occurs.

**Type of Data Science Problem:**

Supervised Learning (since the model will be trained on historical customer data labeled with known outcomes: "churned" or "retained").

**Target:**

Customer churn status (a discrete categorical variable, typically represented as a binary outcome: Yes/True or No/False).


---

5. Comcast wants you to develop a chat bot (automated customer service) to
take text input and respond with the appropriate help and/or service
response.



**Statement:**

Data Science Problem Statement: Process and analyze natural language text inputs from customers to classify the intent of their inquiries and automatically map them to the most appropriate help or service response category.

**Type of Data Science Problem:**

Supervised Learning (since the system relies on human-labeled historical chat logs where text inputs are paired with known, correct intent categories and responses).

**Target:**

The customer inquiry intent or specific help response category (a discrete categorical variable).



---

6. Instagram wants you to analyze images and output whether the image
contains people, animals, scenery, or some other subject.


**Statement:**

Analyze the visual features of uploaded images to automatically categorize each image into one of several distinct subject types: people, animals, scenery, or other subjects.

**Type of Data Science Problem:**

Supervised Learning (since the computer vision model must be trained on a dataset of historical images that have been explicitly labeled with their correct visual categories).

**Target:**

The primary subject category of the image (a discrete categorical variable with options such as "people", "animals", "scenery", or "other").
